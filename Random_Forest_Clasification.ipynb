```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import ConfusionMatrixDisplay
import pickle
import numpy as np
```

```python
file_path = "/content/drive/MyDrive/Celerates MSIB/ALL Dataset/creditcard.csv"
```

```python
# Load Data
df = pd.read_csv(file_path)
print (df.shape)
df.head()
```

Tabel Utama : Kolom Time s.d kolom class


*  Pisahkan X dan Y, dimana : X = fitur atau atribute atau faktor (Semua Kolom selain kolom class), Y =  target (hanya berisi kolom class)
*  Train test split : X dan Y masing-masing akan dibagi 2, X -> (x_train & x_test) Y -> (y_train & y_test)




```python
X = df.drop("Class", axis=1)
Y = df["Class"]
```

```python
# Train Test Split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
print(x_train.shape, x_test.shape)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# Modeling with Rendom Forest
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
# evaluation
y_pred = rf.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
#Visualisasi confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```
